# Método elbow (cotovelo)
## Basicamente testa a variância dos dados em  relação ao número de clusters.

* Até o momento que aumentar o número de 
* Clusters não representa melhoria significativa.



## A medida que a quantidade de clusters aumenta, a soma das distâncias quadráticas tende a zero. 

```
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans


X, y = make_blobs(n_samples=50, n_features=2)

 
erros = [] #somatório dos erros quadráticos das instâncias de cada cluste

for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, init='random')
    kmeans.fit(X)
    print(     i, kmeans.inertia_) 
    # (inertia_ mostra a soma quadrática das distâncias para o centróide mais próximo).
    
    erros.append(kmeans.inertia_)
    
    
plt.plot(range(1, 10), erros)
plt.title('O Metodo Elbow (cotovelo)')
plt.xlabel('Numero de Clusters')
plt.ylabel('Somatório das distâncias quadráticos')  # within cluster sum of squares
plt
```
