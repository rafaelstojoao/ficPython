```
import pandas as pd
from sklearn import tree
import numpy as np
import matplotlib.pyplot as plt

dados = pd.read_csv('../dataset/clima.csv') 
dados.head()

tempMax = dados.iloc[0:1000,1] # valor de saida esperada
dadosTreinamento = dados.iloc[0:1000,2:]

dadosTeste = dados.iloc[1000:,2:]
tempMaxTeste = dados.iloc[1000:,1]



fig, ax = plt.subplots(2,1,figsize=(15,10))
ax[0].scatter(np.arange(len(tempMaxTeste)),tempMaxTeste)
ax[0].set_title("Real")



regressor = tree.DecisionTreeRegressor()
regressor.fit(dadosTreinamento,tempMax)


predicoes = regressor.predict(dadosTeste)

ax[1].scatter(np.arange(len(predicoes)),predicoes)
ax[1].set_title("Predito")

fig.show()


print("Taxa de correlação: \n",regressor.score(dadosTeste,tempMaxTeste))
 #
```

```
from sklearn.metrics import *

real = tempMaxTeste
predito = predicoes

print(max_error(real, predito))
print(mean_absolute_error(real,predito))
print(mean_squared_error(real,predito))
```
