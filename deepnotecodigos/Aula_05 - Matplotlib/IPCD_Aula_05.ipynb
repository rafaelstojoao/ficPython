```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(15,5)) #stateless

# Agora podemos separar nossas informações ou melhor, SEPARAR as visões de para um mesmo conjunto de dados.

#grafico 1
x1 = np.arange(1,100,1)
y1 = np.cos(x1)
axes[0].plot(x1,y1,c='r')


#gráfico 2
x2 = x1
y2 = np.sin(x1)
axes[1].bar(x1,y2)



# A customização continua bem parecida, mas para cada gráfico de forma independente
# axes[0].set_title('Titulo')
# axes[0].set_xlabel('Eixo X')
# axes[0].legend(['Legenda do meu gráfico'])
# axes[0].set_xlim(-10,150)

fig.show()



```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x1, x1,c='y')
axs[0, 0].set_title('Grafico [0,0]')

axs[0, 1].plot(x1, y2, c='orange')
axs[0, 1].set_title('Grafico [0,0]')

axs[1, 0].plot(y1, x2, c='green')
axs[1, 0].set_title('Grafico  [1,0]')

axs[1, 1].plot(y2, y1, c='red')
axs[1, 1].set_title('Grafico  [1,1]')

for ax in axs.flat:
    ax.set(xlabel='Valores de X', ylabel='Valores de Y')

#esconde os x label dos graficos de cima e os y label dos gráficos da direita
for ax in axs.flat:
    ax.label_outer()
```

```python
#https://github.com/plotly/datasets
#https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv

import pandas as pd
dados = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv',)
dados.head()

novoDF = dados[['continent','total_cases','new_cases','total_deaths','new_deaths','population','population_density','median_age','aged_70_older'	]].copy()
# novoDF.head()
novoDF.fillna(0)
# novoDF.head()
# novoDF.to_csv('dadosSelecionados.csv')




```

```python

mediaContinentes = novoDF.groupby('continent').mean()
somaContinentes = novoDF.groupby('continent').sum()

mediaContinentes.head()

# selecao = novoDF.loc[(novoDF['new_cases'] > 100)]
# selecao.head()

```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(15,10))

mediaContinentes.head()
legendas = list(mediaContinentes.index)
legendaNum = np.arange(0,len(legendas))



ax.set_xticks(legendaNum)
ax.set_xticklabels(legendas, rotation='vertical', fontsize=12)

ax.bar(legendaNum,mediaContinentes['population_density'])
ax.bar(legendaNum,mediaContinentes['new_cases'],alpha=0.2)

ax.plot(legendaNum,mediaContinentes['new_deaths'],c='g',marker='*')
ax.legend(['Total de mortes','densidade_populacional','Novos casos'])

fig.show()
```

# Gráfico de dispersão e a correlação entre os valores

## A correlação de pearson é muito utilizada para identificar/indicar a implicação (correlação) linear entre duas variáveis numéricas

* pode ser positiva
* pode ser negativa
* pode ser nula


## pode variar em termos de valor de -1 a +1. Quanto maior for o valor absoluto do coeficiente, mais forte é a relação entre as variáveis.

```python
horasTrabalhadas = np.array([10, 8, 7, 8, 12, 11, 13])
salario = np.array([3500,2500,2000,2700,4000,3800,6500])

print(np.corrcoef(horasTrabalhadas, salario)[0, 1])

```

## mas muitas vezes é difícil acreditar e convencer apenas com um número.
## Visualizar essa correlação facilita a análise



# O ** scatterplot **  do matplotlib é um gráfico que pode ser utilizado para a dispersão dos dados -  considerada uma das principais ferramentas dos cientistas de dados (ao lado do histograma)

```python
plt.scatter(horasTrabalhadas,salario)
plt.show
```

```python
#Com dataframes em pandas é possível fazer um mapa de correlação de forma bem mais simples
mediaContinentes.corr().style.background_gradient(cmap='Blues')
```

```python
# quando uma correlação linear não for identificada  a correlação de Spearman pode contribuir com a análise

from scipy import stats

stats.spearmanr(horasTrabalhadas,salario)
# stats.pearsonr(horasTrabalhadas,salario)


#O p-value é a medida  de significância com que se rejeitaria a hipótese nula. Quanto menor o p-value, mais força a correlação tem
#p acima de 0.05 rejeitam a afirmação
```

```python
from numpy.random import seed
from numpy.random import randn

seed(1)
# prepação:
data1 = 20 * randn(1000) + 100 
data2 = data1 + (10 * randn(1000) + 50) 

# media e desvio padrao de data1
print('média de data1:' '', str(round(np.mean(data1),2)))
print('desvio padrão de data1:' '', str(round(np.std(data1),2)))

# media e desvio padrao de data2
print('média de data2:' '', str(round(np.mean(data2),2)))
print('desvio padrão de data1:' '', str(round(np.std(data2),2)))

# plot
plt.scatter(data1, data2)
plt.title('Gráfico de Dispersão entre data1 e data2')
plt.show()
```
