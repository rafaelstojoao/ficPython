```
import os
import pandas as pd
from matplotlib import pyplot as plt
```

```
# listando todos os arquivos csv
caminhos = [os.path.join(nome) for nome in os.listdir("/home/jovyan/work/dados")]
csvs = [arq for arq in caminhos if arq.lower().endswith(".csv")]
# removendo os arquvios que não me interessam
csvs.remove('VotosMesoMacro202001.csv')
print(csvs)
```

```
#Atribuindo todos para o dataframe
df_dadosEleicao2 = pd.DataFrame()

for arquivo in csvs:
    caminho = ""
    caminho = "/home/jovyan/work/dados/" + arquivo
    df =  pd.read_csv(caminho, encoding = "Latin 1", sep=';')
    df_dadosEleicao2 = df_dadosEleicao2.append(df)     
    #print(caminho)
    #print(df_dadosEleicao2.isna().sum())
```

```
df_EleicaoVotos = df_dadosEleicao2.loc[:, ['NM_MUNICIPIO','QT_VOTOS_NOMINAIS','SG_PARTIDO','QT_VOTOS_LEGENDA', 'NM_COLIGACAO', 'DS_COMPOSICAO_COLIGACAO']]
#df_EleicaoVotos.isna().sum()
```

```
#Inclui uma coluna COMPOSICAO_SPLIT (nova serie), com a composição das coligações  
df_EleicaoVotos['COMPOSICAO_SPLIT'] = df_EleicaoVotos.loc[:, 'DS_COMPOSICAO_COLIGACAO'].str.split('/')
#df_EleicaoVotos.isna().sum()
```

```
#A série como numPy Array
aColigacoes = df_EleicaoVotos['COMPOSICAO_SPLIT'].values
print(type(aColigacoes))
```

```
# Cada coligação agora é uma lista
print(5*'=')
print(type(aColigacoes[9875]))
print(aColigacoes[616])

# e as listas são de strings
print(5*'=')
print(type(aColigacoes[616][0]))
print(aColigacoes[616][0])
print(aColigacoes[616][1])
```
