```
#Hipônimo: palavra que pertence ao mesmo campo semântico mas com sentido mais restrito, ou seja, 
#           é um vocábulo mais específico. Por exemplo: Observar, examinar, olhar, enxergar são hipônimos de ver.
motorcar = wordnet.synset('car.n.01')
types_of_motorcar = motorcar.hyponyms()
types_of_motorcar
```

```
# Listando todos os hipônimos para motorcar
# lemma.name()
[lemma.name() for synset in types_of_motorcar for lemma in synset.lemmas()]
```

```
#buscando o synonym set (synset) para a palavra language
from nltk.corpus import wordnet
wordnet.synsets('language')
```

```
#pos() -> lexycal category
#v - verb n - noun a - adjective adv - adverb prep - preposition num - numbers 
#intj - interjection art - article c - conjunction scon - subordinating conjunction
print(wordnet.synset('language.n.01').pos())
print(wordnet.synset('language.n.01').lemmas())
print(wordnet.synset('language.n.01').lemmas()[0].name())
```

```
# estraindo algumas informações do objeto synsets
poses = {'n': 'noun', 'v': 'verb', 's': 'adj (s)', 'a': 'adj', 'r':'adv'}
for synset in wordnet.synsets('language'):
    print('{}: {}'.format(
        poses[synset.pos()],
        ", ".join([l.name() for l in synset.lemmas()])
    ))
```

# Word Vectors
* Representar as palavras pelo seu contexto (proximidade)
* Vetores n-dimensional sem valores zeros
* Fontes:
    * #https://towardsdatascience.com/learn-nlp-the-stanford-way-lesson-1-3f1844265760
    * Notebook exemplo para entender Word Vectors: https://gist.github.com/aparrish/2f562e3737544cf29aaf1af30362f469

Analogamente podemos aplicar par textos identificando os contextos similares (similar context) analizando a distribuição do contexto:

* It was really cold yesterday.
* It will be really warm today, though.
* It'll be really hot tomorrow!
* Will it be really cool Tuesday?

As palavras cold, warm, hot e cool devem estar relacionadas de alguma forma (próximidade de significado: entre as palavras realy e um dia em partiular - today, tomorrow,...)

Nesse sentido, a palavra significado, indica uma grande lista de todos os contextos ocorridos.
Duas palavras são próximas em significados, se elas compartilham o mesmo contexto.



### Word vectors pela contagem de contexto

#### Ex.: It was the best of times, it was the worst of times

Na tabela abaixo temos  uma coluna para contexto possível e uma linha para cada palavra

Cada célula representa a contagem de ocorrência de uma palavra para cada contexto

Assim: Cada linha é um word vector


![Picture title](Context_count.png)

* Esse procedimento gera uma grande quantidade de células com zero (fora do contexto), o que permite que apliquemos processos de redução dimensional, para otimização no processamento de grandes corpora
* Os parâmetros para a determinação do contexto, podem variar, por exemplo, em vez de pegarmos as palavras imediatamente anterior e posterior, poderíamos pegar duas para trás e duas para frente, ampliando a janela, ou aplicando lemmantization antes da análise


![Picture title](Medium_vector.png)

# Word2Vec:
* Técnica para processamento de linguagem natural. O algoritmo word2vec usa um modelo de rede neural para aprender associações de palavras a partir de um grande corpo de texto.
* Uma vez treinado, esse modelo pode detectar palavras sinônimas ou sugerir palavras adicionais para uma frase parcial
* grande corpus de textos
* Representa cada palavra por um vetor n-dimensional
* Para cada posição no texto, define uma palavra central e palavras de contexto
* Utiliza a similaridade entre vetores para calcular a probabilidade de um contexto dada uma palavra central
* Repete e ajusta a os word vectors para maximizar esta probabilidade

```
#importação de pacotes e módulos
import pandas as pd
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer

#Criando dat frame a partir dos discursos
X_train = pd.DataFrame([part1,part2], columns=['speech'])
X_train
```

```
# passo 1 : TOKENISE
# cada palavra será uma string separada
# mantem as diferenciações (variações) das palavras quanto ao case, suffixo - próximo passo será a normalização
from nltk.tokenize import RegexpTokenizer

#instância para tokens alfanuméricos
tokeniser = RegexpTokenizer(r'\w+')

tokens = tokeniser.tokenize(part1)

# cada palavra é uma 
print(tokens)
print(len(tokens))


```

#### Stemming e Lemmatization
https://towardsdatascience.com/introduction-to-nlp-part-2-difference-between-lemmatisation-and-stemming-3789be1c55bc
* tecnicas para normalização de textos e forma a obter a raiz da palavra

" Stemming usually refers to a crude heuristic process that chops off the ends of words in the hope of achieving this goal correctly most of the time, and often includes the removal of derivational affixes….Stemmers use language-specific rules, but they require less knowledge than a lemmatizer…"
(https://nlp.stanford.edu/IR-book/html/htmledition/stemming-and-lemmatization-1.html)

"Lemmatization usually refers to doing things properly with the use of a vocabulary and morphological analysis of words, normally aiming to remove inflectional endings only and to return the base or dictionary form of a word, which is known as the lemma….a lemmatizer, which needs a complete vocabulary and morphological analysis to correctly lemmatize words…"
*(https://nlp.stanford.edu/IR-book/html/htmledition/stemming-and-lemmatization-1.html)

Steam utiliza regras predefinidas para tranformar a palavra em um stem enquanto lemmatisation utiliza bibliotecas lexicas e de contexto para deivar o lemma. Lemma é um forma de dicionário, então sempre terauma palavra válida.





```
# Passo 2: Normalização
# transformar as palavras em sua forma raiz
# Stemmin e lemmatisation
#lemmatisation: transforma palavra para suas forma de dicionário assim como remove a distinção de case (lowercase tudo)

#WordNetLemmatizer
from nltk.stem import WordNetLemmatizer

# Instância
lemmatiser = WordNetLemmatizer()

#Tudo lowercase e Lemmatization
lemmas = [lemmatiser.lemmatize(token.lower(), pos = 'v') for token in tokens]
print(lemmas)
print(len(lemmas))
```

* Após a normalização temos 66 palavras, mas nem todas as palavras carregam a mesma contribuição 
para o peso no significado  do texto.
* Algumas palavras não são úteis para a mensagem chave. Para isso usamos stop words, que são palavras comuns que proveem pouco ou nenhum valor para o significado ao texto

```
#Passo 3: Remover Stop Words


# corpus de stopwords 
from nltk.corpus import stopwords


# Boa parte das stopwords estão em minusculo
print(len(stopwords.words('english')))
stopwords.words('english')[:5]
```

```
#Remoção das stopwords de acordo com as disponível na nltk
keywords = [lemma for lemma in lemmas if lemma not in stopwords.words('english')]
print(len(keywords))
print(keywords)
```

```
# passo 4: contagem vetorizada
# converter uma coleção de documento de texto para uma matriz de contagem de tokens

#contagem de palavras
{word: keywords.count(word) for word in set(keywords)}
```

```
import gensim.downloader as api
from gensim.models.word2vec import Word2Vec

# Utilizando api para baixar o corpus text8 que é um texto extraido da Wikipedia
#“First 100,000,000 bytes of plain text from Wikipedia”
corpus = api.load('text8')
# criação do modelo word2vec com o corpus como parâmetro
word_vectors = Word2Vec(corpus)
```

```
# encontrando palavras similares a soccer
word_vectors.most_similar('soccer')
```

```
# identificando a palavra que não se encontra no grupo
word_vectors.doesnt_match("soccer tennis basketball bottle".split())
```

![Picture title](Vector_Composition.png)

```
# Operações matemáticas do tipo: King-Man + woman
# Buscando o equivalente de rei para mulher
result = word_vectors.most_similar(positive=['woman', 'king'], negative=['man'])
print("{}: {:.4f}".format(*result[0]))
```

```
# fazendo uma função para Trabalhar com analogias
def analogy(x1, x2, y1):
    result = word_vectors.most_similar(positive=[y1, x2], negative=[x1])
    return result[0][0]
```

```
analogy('brazil', 'brazilian', 'america')
```

```
analogy('europe', 'european', 'spain')
```
