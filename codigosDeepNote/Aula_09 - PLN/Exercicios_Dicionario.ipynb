```
# criando um dicionário {nome: nota}
student_grades = {'John': 'A', 'Mary' : 'C', 'Rob' : 'B'}

# verificando a nota de John
print(student_grades['John'])
```

```
# Quando não encontra a chave, retorna erro 
print(student_grades['Maple'])
```

```
# Quando queremos o input do usuário, pode ser um problma
student_name = input("Please enter student name: ")
print(student_grades[student_name]) 
```

```
# Uma solução é definir um  default no método get do dicionário
default_grade = 'Not Available'
print(student_grades.get('Maple',default_grade))
```

```
population_dict = {'China': 1439, 'India': 1380, 'USA' : 331, 'France': 65, 'Germany' : 83, 'Spain': 46} 

#pede para o usuário
Country_Name = input("Pelase enter Country Name: ")

# Dá erro de Case Sensitive
print(population_dict[Country_Name])
```

```
# Solução1: Põe tudo em minúscula
population_dict = {'china'.lower(): 1439, 'india'.lower(): 1380, 'USA'.lower() : 331, 'France'.lower(): 65, 'Germany'.lower() : 83, 'Spain'.lower(): 46} 

#pede para o usuário e transforma a resposta em tudo em minúscula
Country_Name = input("Pelase enter Country Name: ").lower()

# Agora tudo é lower case
print(population_dict[Country_Name])
```

```
#Solução para tratar erro de digitação

population_dict = {'china'.lower(): 1439, 'india'.lower(): 1380, 'USA'.lower() : 331, 'France'.lower(): 65, 'Germany'.lower() : 83, 'Spain'.lower(): 46} 


while(True):

    #pede para o usuário e transforma a resposta em tudo em minúscula
    Country_Name = input("Pelase enter Country Name: ").lower()

    if Country_Name == 'exit':
        break
    # verificar se encontrou a palavra exata
    if Country_Name in population_dict.keys():           
        print(population_dict[Country_Name])
    else:
        print("Dado não dispónível para: ", Country_Name)
```

```
# usando a difflib
# usada para comparação de arquivos, strings, listas e produzir diferente informações em vários formatos

import difflib

# compara duas sequências
#isjunk : utilizado para junk elements (espaços em branco, recuo de linha, etc...) são aqueles que serão ignorados
#         quando da comparação de dois blocos de texto.
#a e b: strings que queremos comparar
#autojunk : a heurística  que tratar certas seqauencia de cracteres como junk
difflib.SequenceMatcher(isjunk=None, a='', b='', autojunk=True)
```

```
#Comparando chinna e china
from difflib import SequenceMatcher

# criando um objetio SequenceMatcher de comparação de duas strings
check = SequenceMatcher(None, 'chinna', 'china')

# Imprime uma razão de similaridade - escala de 0 a 1
print(check.ratio()) 
```

```
# https://docs.python.org/3/library/difflib.html
from difflib import get_close_matches

print(get_close_matches('chinna', ['china', 'france', 'india', 'usa' ]))
```

```
Country_Name = 'chinna'

#buscando a chave mais próxima no dicionário
maybe_country = get_close_matches(Country_Name, population_dict.keys())

print(population_dict[maybe_country[0]])

```

```
#Solução para erro de digitação

population_dict = {'china'.lower(): 1439, 'india'.lower(): 1380, 'USA'.lower() : 331, 'France'.lower(): 65, 'Germany'.lower() : 83, 'Spain'.lower(): 46} 


while(True):

    #pede para o usuário e transforma a resposta em tudo em minúcual
    Country_Name = input("Pelase enter Country Name: ").lower()

    if Country_Name == 'exit':
        break
    if Country_Name in population_dict.keys():   
        # Achou
        print(population_dict[Country_Name])
    else:
        #Busca por similar
        maybe_country = get_close_matches(Country_Name, population_dict.keys())
        # Não encontrou dado similar
        if maybe_country == []:
            print("Dado não dispónível para: ", Country_Name)
        else:
            # confirmação do usuário
            ans = input('Você quis dizer %s? Digite s ou n.'%maybe_country[0])
            # Usuário concordou com a sugestão
            if ans == 's':
                print(population_dict[maybe_country[0]])
            else:
                print("Dado não disponível, tente outra vez...")
```
