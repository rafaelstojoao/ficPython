```python
import pandas as pd

dados = pd.read_csv('dataset.csv',)
df = dados.iloc[2:10]
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

nomes = np.arange(0,len(df['Name']))

legendas = df['Name']
print(legendas)
ax.set_xticks(nomes)


ax.set_xticklabels(legendas, rotation='vertical', fontsize=12)
ax.bar(nomes,df['Deaths - newly reported in last 7 days']);
```
