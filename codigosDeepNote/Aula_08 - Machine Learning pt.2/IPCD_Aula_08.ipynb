# Biblioteca cluster do pacote scikit Learning  -->  classe Kmeans
## from sklearn.cluster import Kmeans
## KMeans(n_clusters, init, max_iter, algorithm)
## Parâmetros:
- n_clusters número de clusters que o algoritmo vai gerar (padrão: 8)
- Init modo como o algoritmo será inicializado:
   -  k-means++: método padrão, favorece a convergência.
   - random: Inicializa os centróides de forma aleatória
   - ndarray: Especifica um array de valores indicando qual seriam os centróides a serem usados para a inicialização.
-  max_iter: número máximo de iterações (padrão: 300)
-  algorithm: especifica a versão do algoritmo K-Means a ser utilizada. A versão clássica é executada através do valor full.




```
from sklearn.cluster import KMeans
agrupador = KMeans(n_clusters = 3, init = 'random')


dados = [[ -8.69404955,  -5.32605634],
 [  4.39804123,   2.41215428],
 [ -7.86962113 , -4.56544764],
 [ -7.65785816,  -4.71327949],
 [  6.09711654 ,  3.97231732],
 [ -6.73362942  ,-0.70168028],
 [  5.91697593  , 2.03247127],
 [ -7.81438522 , -2.28762661],
 [  5.9677516  ,  2.39567702],
 [  7.03849608  , 0.71852115]] #10 instâncias com 2 atributos quaisquer
```

# Como extrair informações com o agrupamento?


## fit()

- Constrói os grupos (clusters).


```
resultado = agrupador.fit(dados)
print(resultado)

```


## fit_predict()

- Constrói os clusters e indica a qual cluster cada elemento do conjunto (teste) pertence.


```
resultado = agrupador.fit_predict(dados)
print(resultado)
```


## fit_transform()

- executa o K-means para agrupar os dados e retorna uma tabela de distâncias (para cada elemento).

```
resultado = agrupador.transform(dados)
print(resultado)
```

# Avaliação dos resultados
## Forma mais conhecida: Monte Carlo (roleta):
- Um registro qualquer é selecionado e uma classificação quanto ao grupo que deve pertencer é realizada.

```
# De forma rápida...


from sklearn import metrics
from sklearn import datasets
from sklearn.cluster import KMeans

X, y = datasets.load_iris(return_X_y=True)

kmeans = KMeans(n_clusters=3)

result = kmeans.fit_predict(X)

labels = kmeans.labels_
print(labels)
metrics.accuracy_score(y,result)


# avaliando pela roleta:
entradaTeste = [X[4]]
saidaEsperada = y[4]

res  = kmeans.predict(entradaTeste,saidaEsperada)

print("\n Esperado:", saidaEsperada," Obtido:",res)
```

# Identificar um padrão:
## Nada mais é que identificar valores frequentes (que se repetem) em uma base de dados...


- O valor de frequência de um item é chamado SUPORTE

```
def calcularSuporte(dataset,itemBuscado):
    suporte  = 0.0
    contador = 0
    
    for registro in dataset:
        for itemDataset in registro:
            if itemDataset == itemBuscado:
                contador +=1
    suporte = contador/len(dataset)
    return suporte



#supondo um dataset de páginas acessadas por visitantes de um site 
datasetSite = [
            ['index.html','sobre-nos.html','contato.html','produto1.html','produto3.html','produto8.html'],       
            ['produto1.html','produto3.html','produto12.html','produto21.html'],
            ['index.html'],
            ['produto3.html','contato.html','produto8.html','sobre-nos.html'],
            ['index.html','produto10.html','produto12.html','contato.html'],
            ['index.html']
              ]

#cada linha da base de dados significa um(a) visitante (ip) e as páginas que ele(a) visitou



conjuntoPAdroes = {} #dicionário vazio

for linhaDataset in datasetSite:
    for item in linhaDataset:
        #print(str(item) +' suporte: '+str(calcularSuporte(datasetSite,item)))
        conjuntoPAdroes.update({item:calcularSuporte(datasetSite,item)})
        
        

padroesOrdenados = sorted(conjuntoPAdroes.items(), key=lambda indice: indice[1],reverse=True)
padroesOrdenados
```

# Um padrão é identificado quando seu suporte supera um valor pré definido:
## O min_sup (mínimo suporte)

```
minimum_support = 0.5

for sequencia in padroesOrdenados:
    
    if(sequencia[1]>= minimum_support):
        print(str(sequencia)+' é padrao')
    # else:
    #     print(str(sequencia)+' Não é padrao')
```

## Confiança (Conf) 

- é um valor numérico que representa a ‘força’ que a inferência feita pela regra tem na BD e definida como:

Uma Regra de Associação  é dita forte quando apresenta confiança maior que o valor pré-estabelecido pelo usuário, para confiança mínima, min_conf.

Assim, pode-se afirmar que ao identificar X, Y poderá aparecer…
Essa afirmação é dita com confiança de Conf(Ri)%

# Lift (elevação)
Também conhecida como interest – Avalia a dependência entre as partes da regra

R: X ⇒ Y
lift(X ⇒ Y) indica o quão mais frequente Y se torna conforme X aparece

Lift(X ⇒ Y) = Lift(Y ⇒X)


Quanto maior o valor do lift, mais interessante a regra é – indica que X aumenta (“lifts”) a ocorrência de  Y.

# Conviction (convicção)
Lift e confiança são medidas simétricas.
 
Convicção não – respeita a →    
(conv(X → Y)  !=  conv(Y → X))

Convicção é a métrica da implicação. 

Quanto maior a convicção, maior a dependência de Y por X. 
- Conf  == 1   →  conv tende a ∞
